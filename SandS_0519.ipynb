### Euler's formula

http://en.wikipedia.org/wiki/Euler%27s_formula

$$\ \ \ \ \ \ \ e^{jx} = \cos(x) + j\sin(x)$$

公式可以分成兩個部分:

$$\cos(\omega t) = \mathrm{Re}[ e^{j\omega t}] $$

$$\sin(\omega t) = \mathrm{Im}[ e^{j\omega t}] $$

$$ \ \ \ \ e^{j\pi} + 1 = 0 $$

# Sampling Theorem

* 取樣是將訊號從連續時間的類比訊號轉換到離散時間的離散訊號的過程。通常取樣與量化聯合進行，類比訊號由取樣器按照一定時間間隔取樣獲得時間上離散的數位訊號。

* 信號在取樣時，取樣頻率必須不小於有效取樣頻率。

* 有效取樣頻率的規定又稱為奈奎斯特取樣定理 (Nyquist Sampling Theorem)。

$$ T ≥ 2T $$

# Fourier Transform

* **傅立葉變換**是把複雜的週期函數用一系列簡單的正弦、餘弦波之合表示的方法。

* 在信號處理中傅立葉轉換的典型用途是將信號分解成振幅分量和頻率分量。

$$S(f) = \int_{-\infty}^{\infty} s(t) \cdot e^{- i 2\pi f t} dt$$

![http://upload.wikimedia.org/wikipedia/commons/2/2b/Fourier_series_and_transform.gif](http://upload.wikimedia.org/wikipedia/commons/2/2b/Fourier_series_and_transform.gif)



若當輸入的資料為離散時，則使用**離散傅立葉轉換(Discrete Fourier Transform (DFT):**

$$X_k = \sum_{n=0}^{N-1} x_n \cdot e^{-i 2 \pi k n / N}$$

http://en.wikipedia.org/wiki/Fourier_transform
http://en.wikipedia.org/wiki/Discrete_Fourier_transform
http://en.wikipedia.org/wiki/Fourier_series

```python
phs = np.linspace(0, 2*np.pi, 1000)
x = np.cos(10*phs)

plt.plot(x)
```

```python
xf = np.fft.fft(x)

plt.subplot(121)
plt.plot(abs(xf))
plt.title('The magnitude spectrum')

plt.subplot(122)
plt.plot(np.angle(xf))
plt.title('The phase spectrum')

plt.gcf().set_figwidth(10)
```

```python
plt.subplot(121)
plt.plot(abs(xf))
plt.xlim(0,20)
plt.title('The magnitude spectrum')

plt.subplot(122)
plt.plot(np.angle(xf))
plt.xlim(0,100)
plt.title('The phase spectrum')

plt.gcf().set_figwidth(10)
```

```python
plt.subplot(121)
plt.plot(np.real(xf))
plt.title('Real part')

plt.subplot(122)
plt.plot(np.imag(xf))
plt.title('Imaginary part')

plt.gcf().set_figwidth(10)
```

# Convolution

* 連續時間系統為線性且非時變，摺積公式如下：

$$(f * g )(\tau) \stackrel{\mathrm{def}}{=}\ \int_{-\infty}^\infty f(t)\, g(\tau - t)\, d\tau$$


![http://upload.wikimedia.org/wikipedia/commons/6/6a/Convolution_of_box_signal_with_itself2.gif](http://upload.wikimedia.org/wikipedia/commons/6/6a/Convolution_of_box_signal_with_itself2.gif)

![http://upload.wikimedia.org/wikipedia/commons/b/b9/Convolution_of_spiky_function_with_box2.gif](http://upload.wikimedia.org/wikipedia/commons/b/b9/Convolution_of_spiky_function_with_box2.gif)

```python
import pylab as pl

def conv_m(x, nx, h, nh):
    #begining
    nyb = nx[0] + nh[0];
    #ending
    nye = nx[len(x)-1] + nh[len(h)-1];
    ny = np.arange(nyb, nye+1)
    #convolution
    y = np.convolve(x,h)
    
    return y, ny
```

```python
x = [0, 1, 1, 0, 0]
nx = np.arange(-1, 4)
h = [1, 0.5]
nh = np.arange(1, 3)
y, ny = conv_m(x, nx, h, nh)

pl.stem(ny, y)
#stem 畫出離散信號
```
