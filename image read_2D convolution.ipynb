# Image Filter : 2D Convolution

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import lena
from scipy.ndimage.filters import convolve
```

```python
image = lena()
plt.imshow(image, cmap=plt.cm.gray)
```

```python
mat = np.zeros((100,100))
mat[50,50] = 1
plt.subplot(121)
plt.imshow(mat, cmap=plt.cm.gray)

kernel = np.ones((20,20))
c = convolve(mat, kernel)
plt.subplot(122)
plt.imshow(c, cmap=plt.cm.gray)
plt.colorbar()

plt.gcf().set_figwidth(10)
```

```python
kernel_ghost = np.zeros((51, 51))
kernel_ghost[25,25] = 1
kernel_ghost[50,50] = 0.5
plt.subplot(121)
plt.imshow(kernel_ghost, cmap=plt.cm.gray)

plt.subplot(122)
ghost = convolve(image[:,:], kernel_ghost)
plt.imshow(ghost, cmap=plt.cm.gray)
plt.colorbar()

plt.gcf().set_figwidth(10)
```
