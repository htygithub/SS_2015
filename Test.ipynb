```python
from scipy.ndimage.filters import convolve
mat = np.zeros((100,100))
mat[50,50] = 1
kernel = np.zeros((20, 20))
kernel[10, :] = 1
kernel[:, 10] = 1
plt.imshow(kernel, cmap=plt.cm.gray)
```

```python
c = convolve(mat, kernel)
plt.imshow(c, cmap=plt.cm.gray)
plt.colorbar()
```

```python
kernel = np.zeros((20, 20))
kernel[10, :] = 1
kernel[:, 10] = 1
c = convolve(mat, kernel)
plt.subplot(121)
plt.imshow(c, cmap=plt.cm.gray)
plt.colorbar()

a = np.random.rand(64,64)
snow = np.where (a > 0.997, 1, 0)
c = convolve(snow, kernel)
plt.subplot(122)
plt.imshow(c, cmap=plt.cm.gray)
plt.colorbar()

plt.gcf().set_figwidth(10)
```

```python
import scipy.stats as stats
gauss = stats.norm.pdf(np.arange(50), loc=25, scale=10)
plt.plot(gauss)
```

```python
single_dot = np.zeros((101, 101))
single_dot[50, 50] = 1.0
gauss_kernel = nd.gaussian_filter(single_dot, 5)
plt.plot(abs(np.fft.rfft(gauss_kernel[:, 50])), 'o-')
```
