# Phasor arithmetic

```python
t = np.linspace(0, 5*2*np.pi, 300)

y1 = np.sin(t)
y2 = 0.8 * np.sin(t - 0.5*np.pi)

plt.plot(y1)
plt.plot(y2)
```

```python
plt.plot(y1 + y2)
```

### Euler's formula

http://en.wikipedia.org/wiki/Euler%27s_formula

$$\ \ \ \ \ \ \ e^{jx} = \cos(x) + j\sin(x)$$

公式可以分成兩個部分:

$$\cos(\omega t) = \mathrm{Re}[ e^{j\omega t}] $$

$$\sin(\omega t) = \mathrm{Im}[ e^{j\omega t}] $$

$$ \ \ \ \ e^{j\pi} + 1 = 0 $$

# Sampling Theorem

* 取樣是將訊號從連續時間的類比訊號轉換到離散時間的離散訊號的過程。通常取樣與量化聯合進行，類比訊號由取樣器按照一定時間間隔取樣獲得時間上離散的數位訊號。

* 信號在取樣時，取樣頻率必須不小於有效取樣頻率。

* 有效取樣頻率的規定又稱為奈奎斯特取樣定理 (Nyquist Sampling Theorem)。

$$ T ≥ 2T $$

```python
t = np.linspace(0, 2*np.pi, 10)
t0 = np.linspace(0, 2*np.pi, 100)
plt.plot(t, np.sin(t))
plt.plot(t0, np.sin(t0))
```

# Aliasing

```python
phs = np.linspace(0, 10 * 2 * np.pi, 300)
plt.plot(phs, np.sin(phs))
phs = np.linspace(0, 10 * 2 * np.pi, 12)
plt.plot(phs, np.sin(phs), 'o--')
```
