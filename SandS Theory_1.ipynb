## Phasor arithmetic

```python
t = np.linspace(0, 2*np.pi, 300)
y1 = np.sin(5*t)
y2 = 0.8 * np.sin(5*t - 0.5*np.pi)

plt.plot(y1)
plt.plot(y2)
```

```python
plt.plot(y1 + y2)
```

* 同頻率的正弦波相加可得到頻率相同的合成正弦波。

* 相量的和是由旋轉向量進行合成得到的:

## Imaginary Plane

若將平面轉至複數平面上，則x軸為實部，y軸為複部，而我們可以將向量表示為:

http://en.wikipedia.org/wiki/Complex_plane

$$ a \cdot \cos(\omega t) + b \cdot j \cdot \sin(\omega t)$$

![http://www.electronics-tutorials.ws/accircuits/acp75.gif](http://www.electronics-tutorials.ws/accircuits/acp75.gif)

### Euler's formula

$$\ \ \ \ \ \ \ e^{jx} = \cos(x) + j\sin(x)$$

公式可以分成兩個部分:

$$\cos(\omega t) = \mathrm{Re}[ e^{j\omega t}] $$

$$\sin(\omega t) = \mathrm{Im}[ e^{j\omega t}] $$

$$ \ \ \ \ e^{j\pi} + 1 = 0 $$

http://en.wikipedia.org/wiki/Euler%27s_formula

# Sampling Theorem

* 取樣是將訊號從連續時間的類比訊號轉換到離散時間的離散訊號的過程。通常取樣與量化聯合進行，類比訊號由取樣器按照一定時間間隔取樣獲得時間上離散的數位訊號。

* 信號在取樣時，取樣頻率必須不小於有效取樣頻率。

* 有效取樣頻率的規定又稱為奈奎斯特取樣定理 (Nyquist Sampling Theorem)。

$$ T ≥ 2T $$

```python
t = np.linspace(0, 2*np.pi, 10)
t0 = np.linspace(0, 2*np.pi, 100)

plt.plot(t, np.sin(t))
plt.plot(t0, np.sin(t0))
```

```python
phs1 = np.linspace(0, 2*np.pi, 300)
plt.plot(phs1, np.sin(10*phs1))

phs2 = np.linspace(0, 2*np.pi, 12)
plt.plot(phs2, np.sin(10*phs2), 'o--')
```
