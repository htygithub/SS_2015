```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import lena
from scipy.ndimage.filters import convolve
```

```python
image = plt.imread('Lenna.png')
# image = lena() 和直接讀取lenna圖形的差別
# 呼叫的lena為(512,512)灰階圖像
```

# Q1

```python
plt.subplot(131)
plt.imshow(image[:,:,0], cmap=plt.cm.gray)
plt.colorbar()

plt.subplot(132)
plt.imshow(image[:,:,1], cmap=plt.cm.gray)
plt.colorbar()

plt.subplot(133)
plt.imshow(image[:,:,2], cmap=plt.cm.gray)
plt.colorbar()

plt.gcf().set_figwidth(15)
```

# Q2

```python
green_only = image.copy()
green_only[:,:,0] = 0
green_only[:,:,2] = 0
plt.subplot(121)
plt.imshow(green_only)

blue_only = image.copy()
blue_only[:,:,0] = 0
blue_only[:,:,1] = 0
plt.subplot(122)
plt.imshow(blue_only)

plt.gcf().set_figwidth(10)
```

# Q3

```python
mat = np.zeros((100,100))
mat[50,50] = 1
kernel = np.zeros((20, 20))
kernel[10, :] = 1
kernel[:, 10] = 1

plt.subplot(121)
plt.imshow(kernel, cmap=plt.cm.gray)

c = convolve(mat, kernel)
plt.subplot(122)
plt.imshow(c, cmap=plt.cm.gray)
plt.colorbar()

plt.gcf().set_figwidth(10)
```

# Q4

```python
image = lena()
src_image = image[:,:]

kernel_stamp = np.zeros((50,50))
kernel_stamp[np.arange(50), np.arange(50)] = 1.0
stamp = convolve(src_image, kernel_stamp)

plt.subplot(221)
plt.imshow(kernel_stamp, cmap=plt.cm.gray)
plt.subplot(222)
plt.imshow(stamp, cmap=plt.cm.gray)

kernel_stamp = np.zeros((50,50))
kernel_stamp[25,:] = 1.0
kernel_stamp[:,25] = 1.0
stamp = convolve(src_image, kernel_stamp)

plt.subplot(223)
plt.imshow(kernel_stamp, cmap=plt.cm.gray)
plt.subplot(224)
plt.imshow(stamp, cmap=plt.cm.gray)

plt.gcf().set_figheight(5)
```
