```python
%matplotlib inline
img = plt.imread('Lenna.png')
imshow(img)
img.dtype
img.shape
```

```python
subplot(131)
imshow(img[:,:,0])
colorbar()
subplot(132)
imshow(img[:,:,1])
colorbar()
subplot(133)
imshow(img[:,:,2])
colorbar()

gcf().set_figwidth(15) # get current figure
```

```python
subplot(131)
imshow(img[:,:,0], cmap=cm.gray)
colorbar()
subplot(132)
imshow(img[:,:,1], cmap=cm.gray)
colorbar()
subplot(133)
imshow(img[:,:,2], cmap=cm.gray)
colorbar()

gcf().set_figwidth(15)
```

```python
red_only = img.copy()
red_only[:,:,1] = 0
red_only[:,:,2] = 0
subplot(131)
imshow(red_only)

green_only = img.copy()
green_only[:,:,0] = 0
green_only[:,:,2] = 0
subplot(132)
imshow(green_only)

blue_only = img.copy()
blue_only[:,:,0] = 0
blue_only[:,:,1] = 0
subplot(133)
imshow(blue_only)

gcf().set_figwidth(15)
```
