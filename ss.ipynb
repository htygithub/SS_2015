```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import lena
from scipy.ndimage.filters import convolve
import scipy.ndimage as nd
```

```python
src_image = lena()
```

# Q1

```python
plt.subplot(131)
plt.imshow(src_image, cmap=plt.cm.gray)

# gauss_highpass = original_data - gauss_lowpass
plt.subplot(132)
igf = nd.gaussian_filter(src_image, 3)
igf = src_image - igf
plt.imshow(igf, cmap=plt.cm.gray)

plt.subplot(133)
igf = nd.gaussian_filter(src_image, 1)
igf = src_image - igf
plt.imshow(igf, cmap=plt.cm.gray)

plt.gcf().set_figwidth(15)
```
